```python
df = pd.read_csv('/kaggle/input/cardiogoodfitness/CardioGoodFitness.csv')
# calling all the variables in the dataset
columns_age = df['Age']
columns_income =df['Income']
colums_gender = df['Gender']
colums_education = df['Education']
colums_marital_status = df['MaritalStatus']
colums_usage = df['Usage']
colums_fitness =df['Fitness']
colums_miles =df['Miles']
 
# print(columns_age)
# Calculate the absolute frequency using the value_counts() function
absolute_frequencies_income = df['Income'].value_counts().sort_index()
relative_frequencies_income = df['Income'].value_counts(normalize=True).sort_index()

# Calculate the cumulative absolute frequency using the cumsum() function
cumulative_absolute_frequency_income = absolute_frequencies_income.cumsum()

# Calculate the cumulative relative frequency using the cumsum() function
cumulative_relative_frequency_income = relative_frequencies_income.cumsum()

# Print results
print(f"Absolute frequency:\n {absolute_frequencies_income}")
print(f"Relative frequency:\n {relative_frequencies_income}")
print(f"Cummulative absolute frequency:\n{cumulative_absolute_frequency_income}")
print(f"Cummulative relative frequency:\n{cumulative_relative_frequency_income}")
```

# **Mean, Median, Q1, Q3, IQR, and Outliers Of Age**

```python
# Calculate mean
mean_value_age = np.mean(columns_age)

# Calculate median
median_value_age = np.median(columns_age)

# Display the results
print (f"The mean is {mean_value_age}")
print (f"The median is {median_value_age}")

# Calculate the 25th and 75th percentiles
percentile_25 = np.percentile(columns_age, 25)
percentile_75 = np.percentile(columns_age, 75)

# Display the results
print(f"25th Percentile (Q1): {percentile_25}")
print(f"75th Percentile (Q3): {percentile_75}")
q3 = percentile_75
q1 = percentile_25
# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Display the result
print(f"Interquartile Range (IQR): {iqr}")

# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers
outliers = [x for x in columns_age if x < lower_bound or x > upper_bound]

# Display the results
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

# **Covariance Between Age And Income**

```python
# Calculate the covariance matrix
covariance_matrix = np.cov(columns_age, columns_income)

# Extract the covariance between x and y (element at position (0, 1) or (1, 0))
cov_columns_age_and_columns_income = covariance_matrix[0, 1]

# Display the result
print(f"Covariance between Age and income Is: {cov_columns_age_and_columns_income}")
```

# **Interpolation At The Median Point**

```python
# Points where i want to interpolate is the median_value_age
x_interp = median_value_age

# Perform linear interpolation
y_interp = np.interp(x_interp, columns_age, columns_income)

# Display the result
print(f"Interpolated value at x={x_interp}: {y_interp}")

# Plot the original dataset and the interpolated point
plt.plot(columns_age, columns_income, 'o-', label='Original Data')
plt.scatter(x_interp, y_interp, color='red', label='Interpolated Point')
plt.title('Linear Interpolation')
plt.xlabel('columns_age')
plt.ylabel('columns_income')
plt.legend()
plt.show()
```

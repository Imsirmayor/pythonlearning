**Conditional probability Question 1**

**P[(1-2)and(1-4)|1]**

**For 100 iteration**

**Probability of node (1-2) and (1-4) given that node 1 has occurred**

```python
import random

random.seed(1001)

def simulate_conditional_probability_A_given_B(iterations):
    failure_probability_nodes_12_and_14 = 0.00024  # Probability of A failing
    failure_probability_node_1 = 0.2  # Probability of B failing
    nodes_12_and_14_failures = 0
    node_1_failures = 0
    nodes_12_and_14_given_node_1_failures = 0

    for _ in range(iterations):
        nodes_12_and_14_fails = random.random() < failure_probability_nodes_12_and_14
        node_1_fails = random.random() < failure_probability_node_1

        if node_1_fails:
            node_1_failures += 1
            if nodes_12_and_14_fails:
                nodes_12_and_14_given_node_1_failures += 1
                nodes_12_and_14_failures += 1

    # Conditional probability: P(A Fails | B Fails)
    if node_1_failures > 0:
        return nodes_12_and_14_given_node_1_failures / node_1_failures
    else:
        return None

# Running the simulation
iterations = 100
conditional_probability_100_Q1 = simulate_conditional_probability_A_given_B(iterations)
print("Simulated Conditional Probability of nodes_12_and_14 Failure given node_1 Failure:", conditional_probability_100_Q1)
```

**Statistical tests (t-test) to determine if there are significant differences between simulated and theoretically calculated probabilities**

```python
from scipy.stats import ttest_ind

# Simulated data
simulated_probabilities_100_data = [value if value is not None else 0.0 for value in simulated_data_100]
simulated_probabilities_1000_data = [value if value is not None else 0.0 for value in simulated_data_1000]
simulated_probabilities_10000_data = [value if value is not None else 0.0 for value in simulated_data_10000]

# Perform t-test between theoretical and simulated data 
t_stat_100, p_value_100 = ttest_ind(theoretical_probabilities, simulated_probabilities_100_data)
t_stat_1000, p_value_1000 = ttest_ind(theoretical_probabilities, simulated_probabilities_1000_data)
t_stat_10000, p_value_10000 = ttest_ind(theoretical_probabilities, simulated_probabilities_10000_data)

# Display results
print("\nStatistical Analysis Results:")
print(f"T-test (100 iterations): t-statistic = {t_stat_100}, p-value = {p_value_100}")
print(f"T-test (1000 iterations): t-statistic = {t_stat_1000}, p-value = {p_value_1000}")
print(f"T-test (10000 iterations): t-statistic = {t_stat_10000}, p-value = {p_value_10000}")
```
